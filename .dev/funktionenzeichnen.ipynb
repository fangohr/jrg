# Anleitung: Zeichnen von 1d Funktionen ($y=f(x)$)

## 1. Für Jupyter notebook auswaehlen, wie die Funktion gezeichnet wird
Optionen sind: 
`notebook`, `inline` 

```python
# interactive
%matplotlib notebook

# und zeichen library aktievieren
import pylab
import matplotlib.pyplot as plt

# für schickere Farben, seaborn style aktivieren:
import seaborn
seaborn.set_style('darkgrid')
```

## 2. Funktion, die gezeichnet werden soll, als Pythonfunktion definieren. Z.B. $f(x)=2x - 1$:

```python
def f(x):
    return 2*x - 1
```

## 3. Eine Liste mit x und y-Werten berechnen

```python
# Liste, um x-Werte zu speichern
xs = []
# Liste, um y-Werte (=f(x)) zu speichern
ys = []

n = 11   # 11 Punkte
offset = -1 # faengt by x = -1 an
step = 0.2 # erhöht x von Punkt zu Punkt um 0.2
for i in range(n):
    x = i*step + offset
    xs.append(x)
    
    y = f(x)
    ys.append(y)
    
```

## 4. Grafen zeichnen

```python
pylab.plot(xs, ys, 'o-')
pylab.xlabel('x')
pylab.ylabel('f(x) = 3x - 1')

fig, ax = plt.subplots()
ax.plot(xs, ys, 'o-')
```
