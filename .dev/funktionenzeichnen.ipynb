# Anleitung: Zeichnen von 1d Funktionen ($y=f(x)$)

## 1. Für Jupyter notebook auswählen, wie die Funktion gezeichnet wird
Optionen sind: 
- `notebook`: interactive zoomen und plot speichern
- `inline`  : robust (falls `notebook` nicht funktioniert)

```python
# interactive
%matplotlib inline

# und zeichen library aktievieren
import pylab
import matplotlib.pyplot as plt

# für schickere Farben, seaborn style aktivieren:
import seaborn
seaborn.set_style('darkgrid')
# Verfuegbare styles: `darkgrid`, `whitegrid`, `dark`, `white`, und `ticks`

```

## 2. Funktion, die gezeichnet werden soll, als Pythonfunktion definieren. Z.B. $f(x)=2x - 1$:

```python
def f(x):
    return 2*x - 1
```

## 3. Eine Liste mit x und y-Werten berechnen

```python
# Liste, um x-Werte zu speichern
xs = []
# Liste, um y-Werte (=f(x)) zu speichern
ys = []

n = 11   # 11 Punkte
offset = -1 # faengt by x = -1 an
step = 0.2 # erhöht x von Punkt zu Punkt um 0.2
for i in range(n):
    x = i*step + offset
    xs.append(x)
    
    y = f(x)
    ys.append(y)
    
```

## 4. Grafen zeichnen

```python
fig, ax = plt.subplots()
ax.plot(xs, ys, 'o-')
ax.set_xlabel('x')
ax.set_ylabel('f(x) = 3x - 1')
```

## Optional: Plot speichern

```python
ax.set_title("Supergraph")
fig.savefig('plot.pdf')  # als pdf
fig.savefig('plot.png')  # als png
```

## Optional: 2. Graphen zeichnen

```python
def f2(x):
    return x*x - 2
```

```python
ys2 = []
for x in xs:
    ys2.append(f2(x))
    
fig, ax = plt.subplots()
ax.plot(xs, ys, 'o-', label='3x - 1')
ax.plot(xs, ys2, 'g^-', label='x^2 -2')
ax.set_xlabel('x')
ax.legend()


```
